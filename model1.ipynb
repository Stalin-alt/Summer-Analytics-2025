```python

!pip install pandas numpy geopy

import pandas as pd

df = pd.read_csv("dataset.csv")

base_price = 10
alpha = 5

df['Model1_Price'] = df['Occupancy'] / df['Capacity'] * alpha + base_price
print("MODEL 1 OUTPUT:")
print(df[['ID', 'Occupancy', 'Capacity', 'Model1_Price']].head(40))
```

```python
from bokeh.plotting import figure, show, output_notebook
output_notebook()

p1 = figure(title="Model 1: Price vs Occupancy", x_axis_label='Occupancy', y_axis_label='Price')
p1.circle(df['Occupancy'], df['Model1_Price'], size=8, color="navy", alpha=0.6)
show(p1)

```
